##Comparer les groupes obtenus avec L-S-A et ceux du train

```python
import pandas as pd

# --- Ton code existant ---
# Sélectionner uniquement les colonnes numériques (topics)
topic_cols = [col for col in doc_topic_df.columns if col.startswith("Topic_")]

# Trouver le topic dominant
doc_topic_df["Dominant_Topic"] = doc_topic_df[topic_cols].idxmax(axis=1)
doc_topic_df["Max_Weight"] = doc_topic_df[topic_cols].max(axis=1)

# Regrouper les documents selon leur topic dominant
docs_by_topic = doc_topic_df.groupby("Dominant_Topic").indices

# Afficher les indices des documents regroupés par topic
for topic, doc_indices in docs_by_topic.items():
    print(f"\n{topic} → {len(doc_indices)} documents")
    print(list(doc_indices)[:10])  # afficher les 10 premiers docs

# --- Comparaison avec y_train ---
# Regrouper les indices de documents par classe réelle
train_groups = pd.Series(y_train).groupby(y_train).apply(list).to_dict()

# Comparer les groupes LSA avec les groupes train
for topic, lsa_docs in docs_by_topic.items():
    match_found = False
    for label, train_docs in train_groups.items():
        if set(lsa_docs) == set(train_docs):
            print(f"LSA topic {topic} correspond exactement au groupe de classe {label}")
            match_found = True
            break
    if not match_found:
        print(f"LSA topic {topic} ne correspond exactement à aucun groupe du train")
```
