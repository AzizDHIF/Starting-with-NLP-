#Première solution ( tf-idf calculé sur l"ensemble [documents + query])

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
#solution 1(tf_idf globale sur le document + le query)
# --- Corpus de documents ---
documents = [
    "Python is a popular programming language",
    "Machine learning is a branch of artificial intelligence",
    "Deep learning uses neural networks",
    "Python can be used for data science and AI",
    "Natural language processing helps computers understand human language",

    # IT
    "Cybersecurity is a critical field in IT",
    "Cloud computing allows businesses to scale their IT infrastructure",
    "IT professionals often work on software development and system management",

    # Sport
    "Football is the most popular sport in the world",
    "Athletes need regular training to improve their performance",
    "Technology is increasingly used in sports to analyze player performance",

    # Education
    "Online learning platforms make education more accessible",
    "Teachers play a key role in shaping students' education",
    "Artificial intelligence is being used to personalize education"
]


# --- Phrase de requête ---
query = "artificial intelligence is the future "

# --- Étape 1 : Construire la matrice TF-IDF ---
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents + [query])  # on ajoute la requête

# --- Étape 2 : Calculer la similarité cosinus ---
cosine_sim = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])  # comparer requête vs docs

# --- Étape 3 : Trier les documents par pertinence ---
similarities = cosine_sim.flatten()
ranked_indices = similarities.argsort()[::-1]

print("=== Classement des documents ===")
for idx in ranked_indices:
    print(f"Score: {similarities[idx]:.3f} | Document: {documents[idx]}")
```

#Deuxième solution ( tf-idf calculé séparement sur documents puis sur query  )

```python
#solution 2(tf_idf  sur le documenT puis sur le query séparemment)

tfidf_matrix_solution2_documents = vectorizer.fit_transform(documents)  # on ajoute la requête
tfidf_matrix_solution2_query = vectorizer.transform([query])
# --- Étape 2 : Calculer la similarité cosinus ---
cosine_sim_solution2 = cosine_similarity(tfidf_matrix_solution2_documents, tfidf_matrix_solution2_query)  # comparer requête vs docs
# --- Étape 3 : Trier les documents par pertinence ---
similarities_solution2 = cosine_sim_solution2.flatten()
ranked_indices_solution2 = similarities_solution2.argsort()[::-1]

print("=== Classement des documents ===")
for idx in ranked_indices_solution2:
    print(f"Score: {similarities_solution2[idx]:.3f} | Document: {documents[idx]}")
```
